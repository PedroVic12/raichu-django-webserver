```python
import win32com.client as win32
import time
import os 
import pandas as pd 


xlApp = win32.Dispatch('Excel.Application')
xlApp.Visible = False

def createWorkBook(test):
    xlApp.Workbooks.Add()
    xlApp.ActiveWorkbook.SaveAs(os.path.join(os.getcwd(), f'{test}.xlsx'))
    xlApp.ActiveWorkbook.Close(SaveChanges=0)

def load_wb(path,sheet):
    wb = xlApp.Workbooks.Open(path)
    ws = wb.Worksheets(sheet)
    return ws

def saveAutomate(wb):
    wb.RefreshAll()
    time.sleep(1)
    xlApp.CalculateUntilAsyncQueriesDone()
    xlApp.DisplayAlerts = False
    wb.Save()
    time.sleep(1)
    xlApp.Quit()


def read_excel(ws, column, row):
    data = []
    for i in range(1, row):
        data.append([ws.Range(f'{column}{i}').Value])
    return data


# Open the workbook
wb = xlApp.Workbooks.Open(r"C:\Users\PedroVictorRodrigues\OneDrive - CAMORIM SERVICOS MARITIMOS LTDA\Pasta.xlsx")
print(wb)
ws = wb.Worksheets('Planilha1')
read_excel(ws, 'A', 10)

# Write to the workbook
ws.Cells(1, "A").Value = 'hello world'
ws.Cells(1, "B").Value = 'hello world'
ws.Cells(1, "C").Value = 'hello world'
ws.Cells(1, "D").Value = 'hello world'

#write range
ws.Range("A1:C4").Value = "PEDRO V"

ws.Range(
    ws.Cells(6, 6),
    ws.Cells(5, 5)
).Value = "HELLO"

 # Obter o intervalo de dados
last_row = ws.Cells(ws.Rows.Count, 1).End(-4162).Row
last_col = ws.Cells(1, ws.Columns.Count).End(-4159).Column
data_range = ws.Range(ws.Cells(1, 1), ws.Cells(last_row, last_col))
print(f'Numero de linhas: {last_row}  Numero de colunas: {last_col}')

# Ler os dados para uma lista de listas
data = []
for row in range(1, last_row + 1):
    row_data = []
    for col in range(1, last_col + 1):
        cell_value = data_range.Cells(row, col).Value
        row_data.append(cell_value)
    data.append(row_data)

# Converter a lista de listas em um DataFrame
df = pd.DataFrame(data, columns=[f'Coluna{col}' for col in range(1, last_col + 1)])
display(df)


saveAutomate(wb)
```
